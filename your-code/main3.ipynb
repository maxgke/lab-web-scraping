#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'https://www.imdb.com/list/ls009796553/'
````

````python
# your code here
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
title_raw = [soup.find_all(class_ = 'lister-item-header')[i].get_text().split('\n') for i in range(0,10)]
title = [i[2] for i in title_raw]
title
````

````python
year = [i[3] for i in title_raw]
year
````

````python
summary = [soup.find_all('p',{'class':''})[i].get_text().replace("\n", "") for i in range (0,10)]
summary
````

````python
pd.DataFrame(list(map(list, zip(title,year,summary))), columns= ['Title','Year','Summary'])
````
