```python
temp_week = []

table_rows = soup.select('#wt-ext tbody tr')
for elem in table_rows:
    temp_string=elem.find_all('td')[1].getText()
    weather_summary=elem.find_all('td')[2].getText()

    #print(temp_string)
    (temp_max, temp_min) = parse_temp_string(temp_string)
    #print(temp_min)
    tuple1 = (temp_max, temp_min, weather_summary)
    temp_week.append(tuple1)
    
temp_week

```
